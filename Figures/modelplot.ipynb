```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df_bert = pd.read_csv('bert-base-cased-lora-text-classification_.csv')
df_distillbert = pd.read_csv('distilbert-base-uncased-lora-text-classification_.csv')
df_roberta = pd.read_csv('roberta-base-lora-text-classification_.csv')
```

```python
df_bert
```

```python
fig, ax = plt.subplots(figsize=(7,5)) 
ax.plot(df_bert['Unnamed: 0'],df_bert['Accuracy'],label='BERT')
ax.plot(df_distillbert['Unnamed: 0'],df_distillbert['Accuracy'],label='DistillBERT')
ax.plot(df_roberta['Unnamed: 0'],df_roberta['Accuracy'],label='RoBERTa')
ax.set_ylim(0,0.5)
ax.set_xticks([1,2,3])
ax.set_ylabel('Accuracy')
ax.set_xlabel('Epoch')
ax.legend()
ax.set_title('Comparison of Model Accuracy') 
plt.savefig('Model_comparison.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,5)) 
ax.plot(df_bert['Unnamed: 0'],df_bert['Time'],label='BERT')
ax.plot(df_distillbert['Unnamed: 0'],df_distillbert['Time'],label='DistillBERT')
ax.plot(df_roberta['Unnamed: 0'],df_roberta['Time'],label='RoBERTa')
ax.set_ylim(20,70)
ax.set_xticks([1,2,3])
ax.set_ylabel('Time')
ax.set_xlabel('Epoch')
ax.legend()
ax.set_title('Comparison of Model Training Time') 
plt.savefig('Model_comparison_time.png')
plt.show()
```
