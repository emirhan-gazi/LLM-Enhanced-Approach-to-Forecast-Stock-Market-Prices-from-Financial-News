```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
import json
import pandas as pd
```

```python
path = "/content/drive/MyDrive/LLM_FT/Results" #path to the folder containing the results

os.chdir(path) # change the current working directory to the path where the results are stored 
```

```python
folders = [names for names in os.listdir() if not names.endswith('csv') ] # get the names of the folders containing the results
```

```python
for i in range(len(folders)):
  folder_path = path +'/'+folders[i] # path to the folder containing the results
  print(folders[i]) # print the name of the folder
  os.chdir(folder_path) # change the current working directory to the path where the results are stored 
  model = folder_path # name of the model 
  numbers=  [int(i.split('-')[1]) for i in os.listdir() if not i.endswith('s')] # get the numbers of the checkpoints
  last_checkpoint = sorted(numbers,reverse = True)[0] # get the number of the last checkpoint 

  best_path = folder_path + '/checkpoint-' + str(last_checkpoint) #   path to the folder containing the best checkpoint

  os.chdir(best_path) # change the current working directory to the path where the best checkpoint is stored
  trainer_arg = pd.read_json("trainer_state.json") # read the trainer state file 

  epochs = [int(i['epoch']) for i in trainer_arg['log_history'] if len(i)==7 ] # get the epochs
  if type(trainer_arg['log_history'][0]['eval_accuracy']) != dict : # check if the accuracy is a dictionary or not

    acc = [round(i['eval_accuracy'],2 )for i in trainer_arg['log_history'] if len(i)==7 ] # get the accuracy

  else:
    acc = [round(i['eval_accuracy']['accuracy'],2 )for i in trainer_arg['log_history'] if len(i)==7 ] # get the accuracy


  loss = [round(i['eval_loss'],2 )for i in trainer_arg['log_history'] if len(i)==7 ] # get the loss 
  time = [round(i['eval_runtime'],2 )for i in trainer_arg['log_history'] if len(i)==7 ] # get the time

  train_res = pd.DataFrame(index = epochs,columns = ['Accuracy','Loss','Time']) # create a dataframe to store the results
  train_res['Accuracy'] = acc # store the results
  train_res['Loss'] = loss # store the results
  train_res['Time'] = time # store the results

  with open("adapter_config.json") as file: # read the adapter config file
    lora_par = json.load(file) # load the parameters
    lora_res = pd.DataFrame(index = [0],data = {'alpha':lora_par['lora_alpha'],'dropout':lora_par['lora_dropout'],'r':lora_par['r'],'model': model}) # create a dataframe to store the parameters
  print('Done')
  os.chdir(folder_path) # change the current working directory to the path where the results are stored
  train_res.to_csv(f'{model}_.csv')  # save the results
```
